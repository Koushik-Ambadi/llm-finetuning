```python
def normal_column(valid_dataframe, column_name, id_column='file_name'):
    """
    Extracts list-of-dict entries from a parsed column into flat rows,
    tagging each with its associated ID (like file_name).
    Handles missing dictionary keys gracefully.
    """
    extracted_rows = []

    for idx, row in valid_dataframe.iterrows():
        item_id = row[id_column]
        params = row[column_name]

        if isinstance(params, list):
            for param in params:
                if isinstance(param, dict):
                    # Add the ID to the dictionary (non-destructive)
                    param_copy = {**param, id_column: item_id}
                    extracted_rows.append(param_copy)
        else:
            print(f"⚠️ Skipping row {idx}: `{column_name}` is not a list")

    normalized_df = pd.DataFrame(extracted_rows)

    input_count = valid_dataframe[id_column].nunique()
    output_count = normalized_df[id_column].nunique()
    print(f"[NORMALIZED] {output_count} unique `{id_column}` values (from {input_count} in input)")

    return normalized_df
```

```python
import pandas as pd
import re
from collections import Counter

# Example function to extract tags
def extract_html_tags(text):
    if pd.isna(text):
        return []
    return re.findall(r'<\s*([a-zA-Z0-9]+)', str(text))

column_name = 'postcondition'

# Extract tags from all rows and flatten the list
all_tags = df[column_name].dropna().apply(extract_html_tags).explode()

# Count tags
tag_counts = Counter(all_tags)

# Display result
for tag, count in tag_counts.items():
    print(f"{tag}: {count}")


```
