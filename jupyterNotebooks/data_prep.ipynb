```python
import os
import pandas as pd
from pathlib import Path

def read_excel_files_with_filename(folder_path):
    all_dfs = []

    for file in os.listdir(folder_path):
        if file.endswith(".xlsx") or file.endswith(".xls"):
            file_path = os.path.join(folder_path, file)
            df = pd.read_excel(file_path)

            # Add a column with the filename (without extension)
            df['source_file'] = Path(file).stem

            all_dfs.append(df)

    # Combine all dataframes into one (optional)
    combined_df = pd.concat(all_dfs, ignore_index=True)
    return combined_df

# 🔁 Usage Example
req_path = "C://Users//koushik//Desktop//project//data//new_data//requirements"  # 🔁 Change to your folder path
df_requirements = read_excel_files_with_filename(req_path)
tes_path = "C://Users//koushik//Desktop//project//data//new_data//testcases"
df_testcases = read_excel_files_with_filename(tes_path)
```

```python
df_requirements.to_csv("C://Users//koushik//Desktop//project//data//flat_dataset//requirements_dataset.csv", index=False, encoding='utf-8')
df_testcases.to_csv("C://Users//koushik//Desktop//project//data//flat_dataset//testcases_dataset.csv", index=False, encoding='utf-8')
```
