### comparing using sets

```python
import pandas as pd

# --- 1. Function to extract a set from a DataFrame column ---
def extract_values_as_set(df: pd.DataFrame, column: str) -> set:
    values = set()
    for val in df[column].dropna().unique():
        parts = str(val).split(',')
        for part in parts:
            clean = part.strip()
            try:
                values.add(int(clean))  # convert to int if possible
            except ValueError:
                if clean:
                    values.add(clean)   # fallback to string
    return values
```

```python
# --- 2. Sample column lists for df1 and df2 ---
#df_testcases_cols = ['ID', 'Validates']     # example column names in df1
#df_req_cols = ['ID', 'Validated By', 'Decomposed From', 'Satisfied By']       # example column names in df2
df_testcases_cols = ['Validates'] 
df_req_cols = ['ID'] 
```

```python
# --- 3. Extract sets and store with dynamic variable names ---
df_testcases_sets = {}
df_req_sets = {}

for col in df_testcases_cols:
    df_testcases_sets[f'df_testcases_{col.replace(" ", "_")}_set'] = extract_values_as_set(df_testcases, col)

for col in df_req_cols:
    df_req_sets[f'df_req_{col.replace(" ", "_")}_set'] = extract_values_as_set(df_req, col)
```

```python
# --- 4. Compare every set from df1 with every set from df2 ---
for df1_key, set1 in df_testcases_sets.items():
    for df2_key, set2 in df_req_sets.items():
        intersection = set1 & set2
        union = set1 | set2
        only_in_df1 = set1 - set2
        only_in_df2 = set2 - set1

        print(f"🔍 Comparing {df1_key} vs {df2_key}")
        print(f"  → Intersection: {len(intersection)} items")
        print(f"  → Only in {df1_key}: {len(only_in_df1)} items")
        print(f"  → Only in {df2_key}: {len(only_in_df2)} items")
        print(f"  → Jaccard Similarity: {len(intersection)/len(union):.2f}" if union else "  → No union to compare")
        print("—" * 60)
```

### combinig files 

```python
# Step 1: Keep all columns and split Validates by ','
df_validates_expanded = (
    df_testcases
    .assign(Validates=df_testcases['Validates'].apply(lambda v: str(v).split(',') if pd.notna(v) else v))
    .explode('Validates')
)

# Step 2: Merge on cleaned 'Validates' and 'ID' from requirements
df_validates_expanded['Validates'] = df_validates_expanded['Validates'].astype(str)
df_req['ID'] = df_req['ID'].astype(str)

df_joined = df_validates_expanded.merge(
    df_req,
    left_on='Validates',
    right_on='ID',
    how='left',
    suffixes=('.testcase', '.requirement')
)
```
