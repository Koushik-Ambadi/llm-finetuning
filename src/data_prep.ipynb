```python
# Apply across the column
column_name='postcondition'
all_tags = extract_tags_from_column(df, column_name)

# Flatten and count occurrences
flat_tags = [tag for sublist in all_tags for tag in sublist]
tag_counts = Counter(flat_tags)

# Show all unique tags
print("Unique HTML tags found:")
for tag, count in tag_counts.items():
    print(f"<{tag}>: {count} times")
```

```python
from bs4 import BeautifulSoup

def strip_html(text):
    if pd.isna(text):
        return text
    return BeautifulSoup(str(text), "html.parser").get_text()

df[column_name] = df[column_name].apply(strip_html)
df.head(4)
```
