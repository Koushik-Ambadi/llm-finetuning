```python
import ast
df['Parameter'] = df['Parameter'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
df['Parameter']
```

```python
# Step 2: Flatten with file_name
def flatten_params_row(row):
    params = row['Parameter']
    file_name = row['file_name']
    
    if not isinstance(params, list):
        return {'file_name': file_name}
    
    flattened = {'file_name': file_name}
    for param in params:
        if not isinstance(param, dict) or 'no' not in param:
            continue
        prefix = str(param['no']).strip()
        for key, value in param.items():
            if key == 'no':
                continue
            col_name = f"{prefix}_{key}"
            flattened[col_name] = value
    return flattened

# Step 3: Apply to full row
flattened_params = df.apply(flatten_params_row, axis=1)

# Step 4: Convert to DataFrame
flattened_df = pd.DataFrame(flattened_params.tolist())
```
